```python
weather= os.listdir("temp/dataset/")
weather= [f"temp/dataset/{i}" for i in weather]
weather
```

```python
%pwd
```

```python
from tqdm import tqdm

for w in tqdm(weather):
    images= os.listdir(w)
    images_new= [f"{w}/{i}" for i in images]
    i=0
    for image in images_new:
        # print(os.path.exists(image))
        os.rename(
            image,
            f"{w}/weather_{images[i]}"
        )
        i+=1
    os.system(
        f"sudo mv {w}/*.jpg human/"
    )
```

# TU simple US Highways

```python
im_folders= "temp"

for temp_folder in tqdm(os.listdir(im_folders)): # main folder
    temp_folder= f"{im_folders}/{temp_folder}"
    subfolders= os.listdir(temp_folder)
    subfolders= [f"{temp_folder}/{i}" for i in subfolders]
    # print(subfolders)

    for subfolder in subfolders:
        for img in os.listdir(subfolder):
            img_path= f"{subfolder}/{img}"
            new_img_path= f"{subfolder}/highway_{abs(hash(img_path))}.jpg"
            # print(new_img_path)
            os.rename(
                img_path,
                new_img_path
            )
        os.system(
            f"sudo mv {subfolder}/*.jpg human/"
        )
```
